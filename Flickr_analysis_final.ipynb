# Matching Flickr Photos to City Districts in Vienna
Flickr is an easy-to-access data source for user-generated pictures. The pictures contain textual data, e.g. the title, a description or comments. Furthermore, using the transformer package and a huggingface model, it is possible to generate a short description of a picture.

Apart from that, district in Vienna also has associated textual data. This could be for example their wikipedia entries or descriptions of points of interest in each district.

Using spacy and gensim Doc2Vec, we can use all of these textual clues to create an embedding space. Then for each pictures texts the district with the highest similarity associated with the picture. We can verify the experiment by using the geographic coordinates associated with the pictures.

A more complex apporach with using a random subset of the geotagged pictures as training data first is explored at the end.

# Clean text for each picture 
Two functions are provided here. These generate a cleaned version of an input text consisting only of non stop-tokens.

The complete text (for flickr, pois, and wikipedia data) is combined and returned.

```python
def clean_text(text):
    doc = nlp(text)
    alpha_non_stop_tokens = [token.text for token in doc if token.is_alpha and not token.is_stop]
    cleaned_text = ' '.join(alpha_non_stop_tokens)
    return cleaned_text

def make_text(x, with_ai=False):
    text = f"{str(x.title)} {str(x.description)} {str(x.tags)}"
    if with_ai:
        text += x['ai_caption']

    text = clean_text(text)
    
    text = text.replace("wiki", "")
    text = text.replace("Wikipedia", "")
    
    return text
```

```python
def word_sum(x):
    new_word = ""
    for word in x:
        new_word = new_word + " " + word
    return new_word
```

### Clean text for POIs
Now all the textual data of each POI is combined on distict level and cleaned. 

```python
# make sum of text for every district (aggregate rows), using the make_text function from before
wien_en_grouped = wien_locations_eng[["title", "category", "description","zip"]]
wien_en_grouped["tags"] = wien_en_grouped["category"]
wien_en_grouped = wien_en_grouped.groupby("zip").agg(word_sum).reset_index()
wien_en_grouped["sum_text"] = wien_en_grouped.apply(make_text, axis=1)
wien_en_grouped["BEZNR"] = wien_en_grouped["zip"].apply(lambda x: int(str(x)[1:3])/1)
```

```python
vienna_pois = vienna.join(wien_en_grouped[['sum_text', 'BEZNR']], on="BEZNR", rsuffix="_desc").drop(columns=['BEZNR_desc'])
```

### Get info from Wikipedia
A function that uses each district name to get its Wikipedia article and uses the clean_text function from before to prepare the text for vectorization.

```python
def get_info(district):
    district = wikipedia.page(district)
    district_content = district.summary
    district_content = district_content.replace('\n','')

    district_content = clean_text(district_content)
    return district_content
```

## Create the Vectors and get similarities
Our textual model will be used to compare each picture with all of the districts in order to get the most similar district. Our model therefore has to contain the text from all pictures and districts.

```python
# tagged data list
tagged_data = []
```

```python
# tag all data from the district
for idx, row in vienna_pois_desc.iterrows():
    words = str(row['sum_text']) + row['dis_descr']
    tags = int(row['BEZNR'])
    tagged_data.append(TaggedDocument(words=words, tags=['DISTRICT' + str(tags)]))
```

```python
# tag all data from the images
for idx, row in dfSmall.iterrows():
    words = row['all_text']
    tags = row['id']
    tagged_data.append(TaggedDocument(words=words, tags=['PICTURE' + str(tags)]))
```

```python
# build model
model = Doc2Vec(min_count=1, vector_size=100, window=3, workers=3)
model.build_vocab(tagged_data)
model.train(tagged_data, total_examples=model.corpus_count, epochs=10)
```

```python
# get vectors for each district
vienna_with_vectors = vienna_pois_desc
vienna_with_vectors['vector_name'] = vienna_with_vectors['BEZNR'].apply(lambda x: 'DISTRICT'+str(int(x)))
#vienna_with_vectors['vector'] = vienna_with_vectors['vector_name'].apply(lambda x: model.dv[x])
```

```python
# get vectors for each picture
pictures_with_vectors = dfSmall
pictures_with_vectors['vector_name'] = pictures_with_vectors['id'].apply(lambda x: 'PICTURE'+str(int(x)))
#pictures_with_vectors['vector'] = pictures_with_vectors['vector_name'].apply(lambda x: model.dv[x])
```

```python
# function to get most similar district vector
def most_similar_district(pic_vector_name):
    max_sim = -10000
    highest = "NONE_YET"
    
    for name in vienna_with_vectors['vector_name']:
        
        sim = model.dv.similarity(pic_vector_name, name)
        if sim > max_sim:
            
            max_sim = sim
            highest = name
            
    return int(highest[8:])
```

```python
pictures_with_vectors['most_sim_dist'] = pictures_with_vectors.vector_name.apply(most_similar_district)
print("Percentage of correctly coded pictures:")
len(pictures_with_vectors[pictures_with_vectors['correct_district'] == pictures_with_vectors["most_sim_dist"]]) / len(pictures_with_vectors)
```

# Results
We create a confusion matrix using the correct districts and the predicted districts. The confusion matrix is then plotted as a heatmap.
Furthermore we calculate recall, precision and f1-score for each district (restructuring it a two-class problem for each district)

```python
conf = confusion_matrix(pictures_with_vectors['correct_district'], pictures_with_vectors['most_sim_dist'], labels=range(1,24))
conf = conf.astype(float)
conf[conf == 0] = 0.1
```

```python
data = conf

fig, ax = plt.subplots(1, figsize=(12, 12))

# Plot the heatmap
ax.imshow(data, cmap='viridis', interpolation='nearest', norm=LogNorm())

ax.set_xticks(np.arange(len(data[0])), labels=range(1, 24))
ax.set_yticks(np.arange(len(data)), labels=range(1, 24))

ax.tick_params(axis='x', top=True, bottom=False, labeltop=True, labelbottom=False)
ax.set_xlabel('Predicted Labels of Districts')
ax.set_ylabel('True Labels of Districts')

im = ax.imshow(data, cmap='viridis', interpolation='nearest', norm=LogNorm())
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('Number of Pictures')

ax.scatter(range(23), range(23), marker="s", edgecolor="darkred", facecolor="none", s=500, linewidth=3)

for i in range(len(data)):
    for j in range(len(data[0])):
        ax.text(j, i, f'{int(data[i, j])}', color='black',
                ha='center', va='center', fontsize=8)

ax.xaxis.set_label_coords(0.5, 1.07)

fig.savefig("result_matrix_without_flickr_training.png")
```

```python
conf_matrix = pd.DataFrame(columns=["TP","TN", "FP", "FN"])

district_list = [i for i in range(1,24)]
TP_list = []
TN_list = []
FP_list = []
FN_list = []

for district in district_list:
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    for i in range(0, len(pictures_with_vectors)):
        if((district == pictures_with_vectors.iloc[i]["correct_district"]) & (district == pictures_with_vectors.iloc[i]["most_sim_dist"])):
            TP += 1
        if((district != pictures_with_vectors.iloc[i]["correct_district"]) & (district != pictures_with_vectors.iloc[i]["most_sim_dist"])):
            TN += 1
        if((district != pictures_with_vectors.iloc[i]["correct_district"]) & (district == pictures_with_vectors.iloc[i]["most_sim_dist"])):
            FP += 1
        if((district == pictures_with_vectors.iloc[i]["correct_district"]) & (district != pictures_with_vectors.iloc[i]["most_sim_dist"])):
            FN += 1
    TP_list.append(TP)
    TN_list.append(TN)
    FP_list.append(FP)
    FN_list.append(FN)

def calc_f1Score(row):
    f1Score = 0
    if((row["precision"]!=0) & (row["recall"]!=0)):
        f1Score = 2*row["precision"]*row["recall"]/(row["precision"]+row["recall"])
    return f1Score

conf_matrix["district"] = district_list
conf_matrix["TP"] = TP_list
conf_matrix["TN"] = TN_list
conf_matrix["FP"] = FP_list
conf_matrix["FN"] = FN_list
conf_matrix["sum"] = conf_matrix.apply(lambda x: x["TP"]+x["TN"]+x["FP"]+x["FN"], axis=1)
conf_matrix["precision"] = conf_matrix.apply(lambda x: x["TP"]/(x["TP"]+x["FP"]), axis=1)
conf_matrix["recall"] = conf_matrix.apply(lambda x: x["TP"]/(x["TP"]+x["FN"]), axis=1)
conf_matrix["accuracy"] = conf_matrix.apply(lambda x: x["TP"]+x["TN"]/(x["TP"]+x["FP"]+x["TN"]+x["FN"]), axis=1)
conf_matrix["f1Score"] = conf_matrix.apply(calc_f1Score, axis=1)
```

### Create a map for each metric

```python
df_metrics = gpd.GeoDataFrame(conf_matrix, geometry=vienna['geometry'])

def plot_map(colomn):
    fig, ax = plt.subplots(1, figsize=(12, 8))
    df_metrics.plot(column=colomn, ax=ax, legend=True)
    ax.set_axis_off()
    #ax.set_title(colomn)
    fig.savefig("results_no_flickr_training_"+colomn+".png")

plot_map("f1Score")
plot_map("precision")
plot_map("recall")
```

```python
conf_matrix["district"] = district_list
conf_matrix["TP"] = TP_list
conf_matrix["TN"] = TN_list
conf_matrix["FP"] = FP_list
conf_matrix["FN"] = FN_list
conf_matrix["sum"] = conf_matrix.apply(lambda x: x["TP"]+x["TN"]+x["FP"]+x["FN"], axis=1)
conf_matrix["precision"] = conf_matrix.apply(lambda x: x["TP"]/(x["TP"]+x["FP"]), axis=1)
conf_matrix["recall"] = conf_matrix.apply(lambda x: x["TP"]/(x["TP"]+x["FN"]), axis=1)
conf_matrix["accuracy"] = conf_matrix.apply(lambda x: x["TP"]+x["TN"]/(x["TP"]+x["FP"]+x["TN"]+x["FN"]), axis=1)
conf_matrix["f1Score"] = conf_matrix.apply(calc_f1Score, axis=1)
#conf_matrix
```

```python
df_metrics = gpd.GeoDataFrame(conf_matrix, geometry=vienna['geometry'])

def plot_map(colomn):
    fig, ax = plt.subplots(1, figsize=(12, 8))
    df_metrics.plot(column=colomn, ax=ax, legend=True)
    ax.set_axis_off()
    fig.savefig("results_with_flickr_training"+colomn+".png")

plot_map("f1Score")
plot_map("precision")
plot_map("recall")
```
